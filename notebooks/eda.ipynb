```python
## 1. Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## 2. Load Dataset

ata = pd.read_csv('../data/employee_data.csv')

# âœ… Debugging Columns
print("Columns in Data:", data.columns.tolist())
print(data.head())

# âœ… Clean column names (remove any leading/trailing spaces)
data.columns = data.columns.str.strip()

data.head()
```

```python
## 3. Data Summary

data.info()
data.describe()
```

```python
## 4. Missing Values Check

data.isnull().sum()
```

```python
## 5. Salary Distribution

plt.figure(figsize=(8,5))
sns.histplot(data['Salary'], kde=True, color='skyblue')
plt.title('Salary Distribution')
plt.show()
```

```python
## 6. Experience vs Salary

plt.figure(figsize=(8,5))
sns.scatterplot(x='Experience', y='Salary', data=data, hue='Education_Level')
plt.title('Experience vs Salary')
plt.show()
```

```python
## 7. Salary by Job Title

plt.figure(figsize=(8,5))
sns.barplot(x='Job_Title', y='Salary', data=data)
plt.title('Average Salary by Job Title')
plt.xticks(rotation=45)
plt.show()
```

```python
## 8. Salary by Location

plt.figure(figsize=(8,5))
sns.boxplot(x='Location', y='Salary', data=data)
plt.title('Salary by Location')
plt.show()
```

âœ… Conclusion: We visualized salary trends by various features to understand key salary influencers.
